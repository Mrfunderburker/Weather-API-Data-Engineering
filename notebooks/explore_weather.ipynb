```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv
br = pd.read_csv("../data/csv/br_final.csv")
# filter and group data
minas_gerais = br[br["subdivision"] == "Minas Gerais"]
minas_gerais_yearly = minas_gerais.groupby("year")[["million_60kgs_bag"]].sum()
plt.figure(figsize=(10, 6))
# label
sns.lineplot(data=minas_gerais_yearly, x="year", y="million_60kgs_bag", label="60kgs bags of coffee")
plt.title("Harvest Outcomes in Minas Gerais")
plt.xlabel("Year")
plt.ylabel("Bags per million")
```

* How have harvest outcomes changed for Minas Gerais throughout the years?
As the graph illustrates for us, the harvest has had many ups and downs. The data collected shows us that harvests tend to be inconsistent. We can see that the most successful harvest was in what seems to be 2014, with the peak being at around 58 million 60kg bags of coffee. There then seems to be a heavy drop in 2015 when it goes all the way down to 28 million bags. There appears to be a trend where, after every bad year, there is a good one. This can be attributed to many different possibilities. One example could be the health of the soil. After a harvest, the soil may not be as mineral- or nutrient-dense as it was, leading to a worse harvest. As the soil recovers, the harvest production increases. This could also be attributed to the time it takes to grow and prepare for each harvest.

```python
# read csv
br = pd.read_csv("../data/csv/br_final.csv")
# filter and group
regional_trends = br.groupby(["subdivision", "year"])[["million_60kgs_bag"]].sum()
 # plot and label
plt.figure(figsize=(12, 8))
sns.lineplot(data=regional_trends, x="year", y="million_60kgs_bag", hue="subdivision")
plt.title("Harvest Outcomes by subdivision")
plt.xlabel("Year")
plt.ylabel("60kg Bags per Million")
plt.legend(title="Regions")
```

  * Are these harvest outcomes similar to other regions in Brazil? 
The harvest tends to be similar in their own respects. We can see that, overall, other regions in Brazil do not produce as much coffee as Minas Gerais, so we have to compare that data to the region. All regions seem to have had a spike in coffee production in 2014. Though the other regions seem less varied, they also have consistent ups and downs but tend to be less noticeable because of their overall lower production. Espirito Santo does seem to be on a gradual incline, which looks very promising for the area, whereas the other two, Parana and Sao Paulo, seem to be more consistently the same. So, there does seem to be a trend that affects all the regions in Brazil, even if the trend seems less dramatic for others.

```python
# read csv
br = pd.read_csv("../data/csv/br_final.csv")
# filer and group
region_totals = br.groupby(["subdivision", "type"])[["million_60kgs_bag"]].sum().reset_index()
pivot_data = region_totals.pivot(index="subdivision", columns="type", values="million_60kgs_bag")
# plot and label
pivot_data.plot(kind="bar", stacked=True, figsize=(12, 8))
plt.title("Coffee Production by Kind")
plt.xlabel("Region")
plt.ylabel("60kg Bags per MIllion")
plt.legend(title="Type of Coffee")
plt.tight_layout()
```

 * Which region produces the most coffee (by sub-category) in Brazil?
Espirito Santo seems to be the only region in Brazil that grows Robusta coffee. The other regions rely on growing only Arabica. There could be a reason for thismaybe this coffee is more well-known, it tends to do better in the market, or it could also be that this coffee is easier to grow in the region or takes well to the soil. Minas Gerais seems to tower above all the other areas in the production of Arabica coffee, with production almost reaching 600 million 60kg bags, while the leader of Arabica coffee production is Sao Paulo, barely making it to 100 million 60kg bags of coffee. Espirito Santo stands alone with its production of Robusta coffee, seeming to produce more than 200 million 60kg bags of that coffee. Even with its focus on Robusta coffee, Espirito Santo still outperforms Parana with its Arabica coffee production.

```python
# read csv
br = pd.read_csv("../data/csv/br_final.csv")
# filter and group
regional_trends = br.groupby(["subdivision", "year"])[["avg_unemp_perc"]].sum().reset_index()
# plot and label
plt.figure(figsize=(12, 8))
sns.lineplot(data=regional_trends, x="year", y="avg_unemp_perc", hue="subdivision", marker="o")
plt.title("Unemployment Rate Trends by Region")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

* How has unemployment changed in the various coffee-producing regions of Brazil? 
Espirito Santo seems to have by far the highest percentages of unemployment compared to the other regions of Brazil. At its peak, it had more than a 25% unemployment rate. In second place, we have Sao Paulo, reaching almost 15% unemployment at its peak. Following that closely behind is Minas Gerais at around 13%. These three regions had a spike in unemployment after the year 2010, when previously it had been 0%. Parana seems to have had problems with unemployment before the other regions, but after the spike in the other regions, Parana seems to have the lowest rates. The highest rate being 9% in Parana.

```python
# read csv
br = pd.read_csv("../data/csv/combine.csv")
weather_outcome = br.groupby(["year"])[["temperature_2m", "relative_humidity_2m", "precipitation", "surface_pressure"]].sum()
# prepare data
weather_comparison = weather_outcome.loc[weather_outcome.index.isin([2022, 2023])]
weather_comparison.reset_index(inplace=True)
# filter and group
weather_comparison = weather_comparison.melt(id_vars="year", 
value_vars=["temperature_2m", "relative_humidity_2m", "precipitation", "surface_pressure"],
var_name="weather_variable", 
value_name="value")
# plot and label
plt.figure(figsize=(12, 8))
sns.barplot(data=weather_comparison, x="weather_variable", y="value", hue="year", errorbar=None)

plt.title("Weather Comparison Between 2022 and 2023")
plt.xlabel("Weather Variable")
plt.ylabel("Value")
plt.legend(title="Year")
plt.show()
```

* How have weather outcomes changed for Minas Gerais throughout the years?
When comparing the weather conditions in 2022 and 2023, we see that the differnce seems to be relativly the same. The giggest difference we see is when we compare relative humitity. 2023 being only 6 points higher than 2022. temperature also stays almost the same in 2023 only going up by 2 points. precipitions seems to be 0 for both while surface pressure seems to bearly be 1 point higher in 2023.
